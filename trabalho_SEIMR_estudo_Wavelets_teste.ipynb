```python
def so_retina(imagem, mascara):
    image = np.array(imread(imagem),dtype= np.float32)
    image2 = image[:,:,1]
    gray()
    
    mascara_retina = imread(mascara)
    
    mascara_fundo = 255 - mascara_retina  
    
    kernel = zeros_like(image2)
    
    n_linha_f, n_col_f = nonzero(mascara_fundo>0)
    
    n_linha_r, n_col_r = nonzero(mascara_retina>0)
    
    
    for i,j in zip(n_linha_r, n_col_r):
        
        kernel[i,j] = 255 - image2[i,j]
          
    for k,l in zip(n_linha_f, n_col_f):
        
        kernel[k,l] = 0    #255  fundo a branco
        
    return kernel, mascara_retina



def coordenadas():
    angulo = 0
    lista_retas=[]
    
    while angulo <= 165:
        coord = []
        ang = (angulo*math.pi)/180   #passar de graus para radiano
        kernel = np.zeros((15,15))
        b = 7 - tan(ang) * 7         #ordenada na origem descoberta com o ponto que h sempre em comum com todas as retas (7,7)

        if (angulo <= 45) or (angulo >= 135) :
            for j in range(len(kernel[0])):
                i = int(round(tan(ang)*j + b))
                x = len(kernel)-1-i
                coord.append((x -7 ,j -7 ))   #vai guardar as coordenadas em relao ao ponto central da janela


        if (angulo > 45) and (angulo < 135):
            for i in range(len(kernel)):
                j = int(round((i-b)/(tan(ang))))
                
                x = len(kernel)-1-i
                coord.append((x - 7 , j - 7)) 

        lista_retas.append(coord)   
        angulo += 15
    
    return lista_retas



def ortogonais(ang):
    
    ort = {0:'90', 1: '90', 2:'135', 3:'135', 4: '135', 5: '0', 6: '0', 7:'0', 8:'45', 9:'45', 10:' 45', 11:'90'}
    
    lista_cord_ort = []
    
    kernel = zeros((3,3))
    
    angulo_ort = int(ort[ang])              # int() para deixar de ser string
    
    ang_ort = (angulo_ort*pi)/180 
    
    offset = 1 - tan(ang_ort) * 1
    
    if angulo_ort==90:   #todas as linhas
        for i in range (len(kernel)):
            n_col = int(np.around((i-offset)/tan(ang_ort)))
            
            x =len(kernel)- 1 - i
            
            lista_cord_ort.append((x-1,n_col-1))
            
    else:
        for j in range (len(kernel[0])):     #todos ocupam 3 colunas exceto o de 90
            
            n_linha = int(np.around(tan(ang_ort)*j + offset))
            
            x= len(kernel)-1-n_linha
        
            lista_cord_ort.append((x-1,j-1))
    
    return lista_cord_ort




def padding(imagem): 
    n_linha, n_col = imagem.shape
    r = 7
    dim = (n_linha+2*r, n_col+2*r)
    imagem_com_bordas=np.zeros(dim,dtype=(float32))
    #imagem_com_bordas[r:-r, r:-r] = imagem   assim foi como o stor fez
    for i in range (n_linha):
        for j in range (n_col):
            imagem_com_bordas[i+r,j+r]=imagem[i,j]
    return imagem_com_bordas

im, mask = so_retina(imagens_treino[0], mascaras_treino[0])   




def transf_normal(S,S0,I):
    
    norm = []
    lista_feats = [S, S0, I]
    
    for i in range (len(lista_feats)):
        
        med = mean(lista_feats[i])
        dp = np.std(lista_feats[i])
        norma = (lista_feats[i] - med)/dp
        
        norm.append(norma)
            
    return norm     
```

```python
def percorre_imagem_1000p_wv(imag, mascara, wv):     #S, S0 e I
    
    image = imag.copy()
    
    coord= coordenadas()     #12 linhas, cada uma com 15 pixis 

    S_l = []
    S0_l = []
    I_l = []
    wv_cH = []
    wv_cV = []
    wv_cD = []
    
    n_linha, n_col = np.nonzero(mascara>0)  #ler apenas os pixeis   mascara j com o imread()
    
        
    coord_nova=[]
        
    for i in range ((len(n_linha))):       #todas as coordenadas da retina
    
        coord_nova.append((n_linha[i],n_col[i]))
        
        
    #Tirar 1000 coordenadas aleatrias
        
    aleat = sample(coord_nova,1000)      #tirar 1000 coordenadas aleatrias do total de coordenadas da retina

    fim=np.array(aleat)      #1000 coordenadas aleatrias
    
    
    for k in range (len(fim)): #percorrer os 1000 pontos aleatrios
    
        
        i = fim[k,0]
        j = fim[k,1]
        
        centro = (i,j)
    
        
        #1F : 1 PASSO ----- OBTER N - TENDO EM CONTA SE EXISTE ALGUM PIXEL NO FUNDO (IF) OU SE PERTENCEM TODOS  RETINA (ELSE)
            
        
        kernel_imagem = image[i-7:i+8 , j-7:j+8]
        kernel_mask= mascara[i-7:i+8 , j-7:j+8]
        
        
        if (mascara[i-7][j-7] == 0) or (mascara[i-7][j+7] == 0) or (mascara[i+7][j-7] == 0) or (mascara[i+7,j+7] == 0):   #se cumprir tem obrigatoriamente pelo menos um ponto fora da retina
            
            med = mean(kernel_imagem[kernel_mask>0])
            kernel_imagem[kernel_mask==0] = med
            N = mean(kernel_imagem) 
        
        else:
             
            N = mean(kernel_imagem)
        

        #1F : 2 PASSO ----- OBTER L
        
        
        l=0    
        lista_L=[]                      #guarda os valores de L de todas as retas para depois se escolher o mximo
        
        
        while l<=11:                        #percorrer todos os 12 ngulos
                
            lista_ang=[]
            
            
            for k in range (len(coord[0])):   #percorrer todos os 15 pixis para construir as retas de cada ngulo 
                
                pos = coord[l][k]           #obter as coordenadas do pxel nmero k do ngulo l
                    
                novo_pix = (7 + pos[0] , 7 + pos[1])   #obter a posio do pxel da reta em relao ao central------  dava centro[i] e centro[j]?
                
                valor = kernel_imagem[novo_pix[0]][novo_pix[1]]    #usar a imagem principal!
                
                lista_ang.append(valor)
            
            L = mean(lista_ang)
                
            lista_L.append(L)
            
            l+=1
            
        L = max(lista_L)                    #guardar o elemento da lista com maior L
        angulo= lista_L.index(L)            #encontrar a posio do maior L, que vai corresponder ao ngulo-- linha com a coordenadas do ngulo com maior mdia dos 12 da lista
            
        #1F: 3 PASSO ---- OBTER S
            
        S = L - N
            
        #2F: 1 PASSO ----- OBTER L0
            
        coord_ort = ortogonais(angulo)
        
        lista_L0 = []
        
        for p in range (len(coord_ort)):
                
            novo_pix_ort = (7 + coord_ort[p][0], 7 + coord_ort[p][1])
            
            #print(novo_pix_ort)
                
            lista_L0.append(kernel_imagem[novo_pix_ort[0]][novo_pix_ort[1]])
                
        
        L0 = mean(lista_L0)
        
        #2F : 2 PASSO ----- OBTER S0
            
        S0 = L0 - N
        
        #3F : 1 PASSO ---- OBTER I
            
        I = image[i][j]
    
        S_l.append(S)
        S0_l.append(S0)
        I_l.append(I)
        
        wv_cH.append(wv[0][1][0][i][j])
        wv_cV.append(wv[0][1][1][i][j])
        wv_cD.append(wv[0][1][2][i][j])
    
    
    return S_l, S0_l, I_l, fim, wv_cH, wv_cV, wv_cD    #lista_S, lista_S0, lista_I 
```

```python
#Funo para fazer o treino do classificador

from sklearn.svm import SVC        
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GroupKFold

def train_classifier (clf, X_tr, L_tr, L_g_tr):
                                                 
    #TREINO DO CLASSIFICADOR
    
    train_acc_total=0
    test_acc_total=0
    train_sens_total=0
    test_sens_total=0
    train_speci_total=0
    test_speci_total=0
    train_preci_total=0
    test_preci_total=0

    group_kfold = GroupKFold(n_splits=10)

    for train_index, test_index in group_kfold.split(X_tr, L_tr, L_g_tr):
        #print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X_tr[train_index], X_tr[test_index]
        y_train, y_test = L_tr[train_index], L_tr[test_index]
    
        clf = clf.fit(X_train, y_train)
    
        y_pred_train = clf.predict(X_train)
        y_pred_test = clf.predict(X_test)
    
        train_acc = np.mean(y_pred_train == y_train) * 100 
        test_acc= np.mean(y_pred_test == y_test) * 100
    
        cm_train = confusion_matrix(y_train, y_pred_train)
        cm_test = confusion_matrix(y_test, y_pred_test)
    
        TP_train = cm_train[1,1]
        TN_train = cm_train[0,0]
        FP_train = cm_train[0,1]
        FN_train = cm_train[1,0]
    
        TP_test = cm_test[1,1]
        TN_test = cm_test[0,0]
        FP_test = cm_test[0,1]
        FN_test = cm_test[1,0]
    
        train_sens = (TP_train/(TP_train + FN_train))*100 
        test_sens = (TP_test/(TP_test+ FN_test))*100
    
        train_speci = (TN_train/(TN_train + FP_train))*100
        test_speci = (TN_test/(TN_test + FP_test))*100
    
        train_preci = (TP_train/(TP_train+FP_train))*100
        test_preci = (TP_test/(TP_test+FP_test))*100
    
    
    
        train_acc_total+=train_acc
        test_acc_total+=test_acc
    
        train_sens_total+=train_sens
        test_sens_total+=test_sens
    
        train_speci_total+=train_speci
        test_speci_total+=test_speci
    
        train_preci_total+=train_preci
        test_preci_total+=test_preci

    print("Taxa de sucesso (Treino): ",
      (train_acc_total/10))
    print("Taxa de sucesso (Validao): ",
      (test_acc_total/10))
    print("Sensibilidade (Treino): ",
      (train_sens_total/10))
    print("Sensibilidade (Validao): ",
      (test_sens_total/10))
    print("Especificiadade (Treino): ",
      (train_speci_total/10))
    print("Especificidade (Validao): ",
      (test_speci_total/10))
    print("Preciso (Treino): ",
      (train_preci_total/10))
    print("Preciso (Validao): ",
      (test_preci_total/10))
    
    #TESTE DO CLASSIFICADOR
    
    X_t, y_teste, y_group_test = treino_pontos_wv(imagens_treino, mascaras_treino, labels_treino) #TIRAR 1000 PONTOS ALEATRIOS DE CADA IMAGEM DE TREINO

    y_pred = clf.predict(X_t)

    teste_acc = np.mean(y_pred == y_teste) * 100 
    
    cm_teste = confusion_matrix(y_teste, y_pred)
    
    TP_teste = cm_teste[1,1]
    TN_teste = cm_teste[0,0]
    FP_teste = cm_teste[0,1]
    FN_teste = cm_teste[1,0]
    
    teste_sens = (TP_teste/(TP_teste + FN_teste))*100
    
    teste_speci = (TN_teste/(TN_teste + FP_teste))*100
    
    teste_preci = (TP_teste/(TP_teste+FP_teste))*100
    
    print('-------------------------------------------------------------------')
    
    print("Taxa de sucesso (Teste): ",
      (teste_acc))
    print("Sensibilidade (Teste): ",
      (teste_sens))
    print("Especificiadade (Teste): ",
      (teste_speci))
    print("Preciso (Teste): ",
      (teste_preci))
    
    
    return clf
```

```python
def percorre_imagem_teste_wv(imag, mascara,clf, label_teste, wv):     #S, S0 e I
    
    image = imag.copy()
    
    coord= coordenadas()     #12 linhas, cada uma com 15 pixis 

    S_l = []
    S0_l = []
    I_l = []
    wv_cH = []
    wv_cV = []
    wv_cD = []
    
    n_linha, n_col = np.nonzero(mascara>0)  #ler apenas os pixeis   mascara j com o imread()
    
    im_pred = zeros_like(image)
    

    
    for i,j in zip(n_linha,n_col):
            
        centro = (i,j)
        
        #1F : 1 PASSO ----- OBTER N - TENDO EM CONTA SE EXISTE ALGUM PIXEL NO FUNDO (IF) OU SE PERTENCEM TODOS  RETINA (ELSE)
            
        
        kernel_imagem = image[i-7:i+8 , j-7:j+8]
        kernel_mask= mascara[i-7:i+8 , j-7:j+8]

        if (mascara[i-7][j-7] == 0) or (mascara[i-7][j+7] == 0) or (mascara[i+7][j-7] == 0) or (mascara[i+7,j+7] == 0):   #se cumprir tem obrigatoriamente pelo menos um ponto fora da retina
            
            med = mean(kernel_imagem[kernel_mask>0])
            kernel_imagem[kernel_mask==0] = med        
            N = mean(kernel_imagem) 
            
        else:
             
            N = mean(kernel_imagem)
        
        #1F : 2 PASSO ----- OBTER L
        
        l=0
        lista_L=[]                      #guarda os valores de L de todas as retas para depois se escolher o mximo
        
        while l<=11:                        #percorrer todos os 12 ngulos
                
            lista_ang=[]
            
            
            for k in range (len(coord[0])):   #percorrer todos os 15 pixis para construir as retas de cada ngulo 
                
                pos = coord[l][k]           #obter as coordenadas do pxel nmero k do ngulo l    
                novo_pix = (7 + pos[0] , 7 + pos[1])   #obter a posio do pxel da reta em relao ao central------  dava centro[i] e centro[j]?
                
                valor = kernel_imagem[novo_pix[0]][novo_pix[1]]    #usar a imagem principal!
                
                lista_ang.append(valor)
                        
            L = mean(lista_ang)    
            lista_L.append(L)
            l+=1
            
        L = max(lista_L)                    #guardar o elemento da lista com maior L
        angulo= lista_L.index(L)            #encontrar a posio do maior L, que vai corresponder ao ngulo-- linha com a coordenadas do ngulo com maior mdia dos 12 da lista
                                  
        #1F: 3 PASSO ---- OBTER S
            
        S = L - N
        
        #2F: 1 PASSO ----- OBTER L0
            
        coord_ort = ortogonais(angulo)
        
        lista_L0 = []
        
        for p in range (len(coord_ort)):
            
                
            novo_pix_ort = (7 + coord_ort[p][0], 7 + coord_ort[p][1])    
            lista_L0.append(kernel_imagem[novo_pix_ort[0]][novo_pix_ort[1]])
                
        L0 = mean(lista_L0)
        
        #2F : 2 PASSO ----- OBTER S0
            
        S0 = L0 - N
        
        #3F : 1 PASSO ---- OBTER I
        I = image[i][j]
        
            
    
        S_l.append(S)
        S0_l.append(S0)
        I_l.append(I)

        wv_cH.append(wv[0][1][0][i][j])
        wv_cV.append(wv[0][1][1][i][j])
        wv_cD.append(wv[0][1][2][i][j])
    
    
    # Normalizao das 6 features extradas
    
    norma = transf_normal(S_l, S0_l, I_l)
    norma_wv = transf_normal(wv_cH, wv_cV, wv_cD)
    
    
    TP=0
    TN=0
    FP=0
    FN=0
    
    a=-1
    
    for i,j in zip(n_linha,n_col):
        
        a+=1
        
        im_pred[i,j] = clf.predict([[norma[0][a], norma[1][a], norma[2][a], norma_wv[0][a], norma_wv[1][a], norma_wv[2][a]]])*255
        
        if im_pred[i,j] > 0:
        
            if (int(im_pred[i,j]) == int(label_teste[i,j])):
            
                TP +=1
            
            elif (int(im_pred[i,j]) != int(label_teste[i,j])):
            
                FP +=1
        
        else:
        
            if (int(im_pred[i,j]) == int(label_teste[i,j])):
            
                TN +=1
            
            
            elif (int(im_pred[i,j]) != int(label_teste[i,j])):
            
                FN +=1
            
        #print(TP)
            
    Accuracy = ((TP+TN)/(TP+TN+FP+FN)) *100
    
    Precisao = (TP/(TP+FP)) *100
    
    Sensibilidade = (TP/(TP+FN))*100
    
    Especificidade = (TN/(TN+FP))*100
    
    
    return im_pred, Accuracy, Precisao, Sensibilidade, Especificidade
```
